```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/shipments.csv')
print(df.head())
print(f"\nВсего записей: {len(df)}")
```

```python
# Простая визуализация
df.groupby('carrier')['cost_rub'].sum().plot(kind='bar', title='Общая стоимость по перевозчикам')
plt.ylabel('Стоимость, руб')
plt.tight_layout()
plt.show()
```
